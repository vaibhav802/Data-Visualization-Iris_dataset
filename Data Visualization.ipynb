### Write a Python program to create a plot to get a general Statistics of Iris data

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Iris.csv")
df.describe().plot(kind = "area" , fontsize = 15, figsize = (15,8),table = True)
plt.xlabel("Statistics")
plt.ylabel('Value')
plt.title("General statistics of Data")
plt.show()
```

### Write a Python program to create a Bar plot to get the frequency of the three species of the Iris data

```python
df['Species'].value_counts().plot(kind = 'bar', color = ['blue','orange','green'])
plt.xticks(rotation = 0)
plt.show()
```

```python
ax = plt.subplots(1,1,figsize = (10,8))
sns.countplot('Species', data = df)
plt.show()
```

### Write a Python program to create a Pie plot to get the frequency of the three species of the Iris data

```python
ax = plt.subplots(1,1,figsize = (10,8))
df['Species'].value_counts().plot(kind = 'pie',explode=[0.1,0.1,0.1], autopct = '%1.1f%%', shadow = True, labeldistance=1.3)
plt.show()
```

###  Write a Python program to create a graph to find relationship between the sepal length and width

```python
fig = df[df['Species']=='Iris-setosa'].plot(kind = 'scatter', x = 'SepalLengthCm', y = 'SepalWidthCm', color = 'green', label = 'setosa')
df[df['Species']=='Iris-virginica'].plot(kind = 'scatter', x = 'SepalLengthCm', y = 'SepalWidthCm', color = 'red', label = 'virginica',ax= fig)
df[df['Species']=='Iris-versicolor'].plot(kind = 'scatter', x = 'SepalLengthCm', y = 'SepalWidthCm', color = 'blue', label = 'versicolor', ax = fig)
#fig.set_xlabel("Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length VS Width")
#fig.set_ylabel("Sepal Width")
#fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

###  Write a Python program to create a graph to find relationship between the petal length and width

```python
fig = df[df['Species']=='Iris-setosa'].plot(kind = 'scatter', x = 'PetalLengthCm', y = 'PetalWidthCm', color = 'green', label = 'setosa')
df[df['Species']=='Iris-virginica'].plot(kind = 'scatter', x = 'PetalLengthCm', y = 'PetalWidthCm', color = 'red', label = 'virginica',ax= fig)
df[df['Species']=='Iris-versicolor'].plot(kind = 'scatter', x = 'PetalLengthCm', y = 'PetalWidthCm', color = 'blue', label = 'versicolor', ax = fig)
#fig.set_xlabel("Sepal Length")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length VS Width")
#fig.set_ylabel("Sepal Width")
#fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

### Write a Python program to create a joinplot to describe individual distributions on the same plot between Sepal length and Sepal width. 
### Note: joinplot - Draw a plot of two variables with bivariate and univariate graphs.

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", data = df , color = "green")
plt.show()
```

###  Write a Python program to create a joinplot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width.

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", kind = "hex", data = df , color = "blue")
plt.show()
```

### Write a Python program to create a joinplot using "kde" to describe individual distributions on the same plot between Sepal length and Sepal width

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", kind = "kde", data = df , color = "green")
plt.show()
```

### Write a Python program to create a joinplot and add regression and kernel density fits using "reg" to describe individual distributions on the same plot between Sepal length and Sepal width.

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", kind = "reg", data = df)
plt.show()
```

### Write a Python program to draw a scatterplot, then add a joint density estimate to describe individual distributions on the same plot between Sepal length and Sepal width

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", data = df).plot_joint(sns.kdeplot, n_levels =6)
plt.show()
```

### Write a Python program to create a joinplot using "kde" to describe individual distributions on the same plot between Sepal length and Sepal width and use '+' sign as marker

```python
sns.jointplot(x = "SepalLengthCm", y = "SepalWidthCm", kind = "kde", data = df).plot_joint(plt.scatter, marker = '+', c = 'b')
plt.show()
```

###  Write a Python program using seaborn to Create a kde (Kernel Density Estimate ) plot of sepal_length versus sepal width for setosa species of flower

```python
sns.kdeplot(data = df[df['Species']=="Iris-setosa"][["SepalLengthCm","SepalWidthCm"]], color = "blue",shade = True, shade_lowest= False)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

### Write a Python program using seaborn to Create a kde (Kernel Density Estimate ) plot of petal_length versus petal width for setosa species of flower.

```python
sns.kdeplot(data = df[df['Species']=="Iris-setosa"][["PetalLengthCm","PetalWidthCm"]], color = "blue",shade = True, shade_lowest= False)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

### Write a Python program using seaborne to create a kde (Kernel Density Estimate) plot of two shaded bivariate densities of Sepal Width and Sepal Length.

```python
plt.subplots(figsize=(10, 8))
sns.kdeplot(data=df[['SepalLengthCm','SepalWidthCm']],cmap="plasma", shade=True, shade_lowest=False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')
plt.show()
```
